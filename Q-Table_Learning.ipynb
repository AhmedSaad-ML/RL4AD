```python
"""
This code implement simple RL using Q-Table
Source 
https://medium.com/emergent-future/simple-reinforcement-learning-with-tensorflow-part-0-q-learning-with-tables-and-neural-networks-d195264329d0
"""
```

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the environment
env = gym.make('FrozenLake-v0')
```

```python
# Initializing the Q-Table
Q = np.zeros([env.observation_space.n,env.action_space.n])
```

```python
# Set learning parameters
lr = 0.8    # learning rate
g  = 0.95   # discount for future rewards
n  = 2000   # number of episodes
```

```python
# List containing rewards
rList = []
```

```python
for i in range(n):
    s = env.reset()        # resetting the environment
    rAll = 0               # setting total rewards to zero
    d = False              # ???
    j = 0
    # The Q-Table learning algorithm
    while j < 99:
        j += 1
        # Choose and action from Q table by greedy method + noise
        a = np.argmax(Q[s,:] + np.random.randn(1,env.action_space.n)*(1./(i+1)))
        # Get new state and reward from environment
        s1,r,d,_ = env.step(a)
        # Update Q-Table with new knowledge
        Q[s,a] = Q[s,a] + lr*(r + g*np.max(Q[s1,:]) - Q[s,a])
        # Update the total reward
        rAll += r
        # Change the state to new state
        s = s1
        # Discontinue if d = True ??
        if d == True:
            break
    rList.append(rAll)
    #print('------------------------------------')
    #print('Intermediate results')
    print('i, rAll : ' + str(i) + ',' + str(rAll))
    #print('Q-Table : ')
    #print(Q)
    #print('------------------------------------')
        
```

```python
plt.plot(rList[0:1000])
```

```python
plt.show()
```

```python
print "Score over time: " +  str(sum(rList)/n)
```
